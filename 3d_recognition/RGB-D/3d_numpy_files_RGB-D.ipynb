```python
import gzip
import cv2 as cv
import os
import numpy as np
import random
```

```python
def write_on_disk(data, name):
    
    if data.shape[0] > 0:
    
        print("----Writing on Disk: "+str(name)+"----")

        with gzip.GzipFile(np_path+name, "w") as f:
            np.save(f, data)
```

```python
def extract_data(data_path):
    train_labels = []
    test_labels = []
    label_names = []
    IMG_SIZE = 128
    IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
    data_np = np.empty((0,128,128,3), float)


#     vgg = tf.contrib.keras.applications.VGG16(input_shape=IMG_SHAPE, weights="imagenet", include_top=False)
#     vgg.trainable = False
    
    print("\n..........loading dataset from disk..........\n")

    for train_or_test in os.listdir(data_path):
        
        projections_path = os.path.join(data_path, train_or_test)

        projections = os.listdir(projections_path)
        
        labels_path = os.path.join(projections_path, projections[0])

        labels =  os.listdir(labels_path)
        
        count = 0
        n = 0
        for label in labels:

            images_names =  os.listdir(os.path.join(labels_path, label))

            for i in images_names:
                image1 = cv.imread(data_path+"/"+str(train_or_test)+"/"+str(projections[0])+"/"+str(label)+"/"+str(i), cv.IMREAD_GRAYSCALE)
                image2 = cv.imread(data_path+"/"+str(train_or_test)+"/"+str(projections[1])+"/"+str(label)+"/"+str(i), cv.IMREAD_GRAYSCALE)
                image3 = cv.imread(data_path+"/"+str(train_or_test)+"/"+str(projections[2])+"/"+str(label)+"/"+str(i), cv.IMREAD_GRAYSCALE)
                image = cv.merge([image1, image2, image3])
                
                data = np.array([image], dtype="float") / 255.0
                data_np = np.append(data_np, data, axis=0)
                count = count + 1
#                 print(count)
                if count >= 500:
                    n = n + 1
                    write_on_disk(data_np, name=str(train_or_test)+str(n)+".npy.gz")
                    data_np = None
                    data_np = np.empty((0,128,128,3), float)
                    count = 0

                if str(train_or_test) == "train":
                    train_labels.append(label)
                                        
                elif str(train_or_test) == "test":
                    test_labels.append(label)
        
        write_on_disk(data_np, name=str(train_or_test)+str(n+1)+".npy.gz")
        data_np = None
        data_np = np.empty((0,128,128,3), float)
    
    train_labels = np.array(train_labels)

    test_labels = np.array(test_labels)

    label_names = np.unique(train_labels)

    return train_labels, test_labels, label_names
```

```python
if __name__ == "__main__":
    
    data_path = '../../RGB-D/image_dataset'
    
    np_path = '../../RGB-D/numpy_files/'
    np_train_label_path = '../../RGB-D/numpy_files/train_labels.npy'
    np_test_label_path = '../../RGB-D/numpy_files/test_labels.npy'
    np_label_names_path = '../../RGB-D/numpy_files/label_names.npy'

    os.makedirs(os.path.dirname(np_path), exist_ok=True)
    
    train_labels, test_labels, label_names = extract_data(data_path)

    np.save(np_train_label_path, train_labels)
    np.save(np_test_label_path, test_labels)
    np.save(np_label_names_path, label_names)
```
